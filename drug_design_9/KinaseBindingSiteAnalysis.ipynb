```python
import requests
import time
from requests.exceptions import HTTPError

def submit_dogsitescorer_job_with_pdbid(pdb_code, chain_id, ligand="", max_retries=3, retry_delay=60):
    """
    Submit PDB ID to DoGSiteScorer webserver using their API and get back URL for job location.
    Implements retry logic with exponential backoff.
    
    Parameters
    ----------
    pdb_code : str
        4-letter valid PDB ID.
    chain_id : str
        Chain ID to analyze (e.g., 'A').
    ligand : str
        Optional: Name of ligand bound to PDB structure.
    max_retries : int
        Maximum number of retry attempts.
    retry_delay : int
        Initial delay between retries in seconds.
    
    Returns
    -------
    str
        Job location URL.
    """
    for attempt in range(max_retries):
        try:
            r = requests.post(
                "https://proteins.plus/api/dogsite_rest",
                json={
                    "dogsite": {
                        "pdbCode": pdb_code,
                        "analysisDetail": "1",
                        "bindingSitePredictionGranularity": "1",
                        "ligand": ligand,
                        "chain": chain_id,
                    }
                },
                headers={"Content-type": "application/json", "Accept": "application/json"},
            )
            
            r.raise_for_status()
            return r.json()["location"]
        
        except HTTPError as e:
            if e.response.status_code == 429 and attempt < max_retries - 1:
                wait_time = retry_delay * (2 ** attempt)  # Exponential backoff
                print(f"Rate limit exceeded. Waiting for {wait_time} seconds before retrying...")
                time.sleep(wait_time)
            else:
                raise

    raise Exception("Max retries exceeded")
```

```python
import requests
import time
from requests.exceptions import HTTPError

def submit_dogsitescorer_job_with_pdbid(pdb_code, chain_id, ligand="", max_retries=5, initial_retry_delay=60):
    """
    Submit PDB ID to DoGSiteScorer webserver using their API and get back URL for job location.
    Implements retry logic with exponential backoff.
    
    Parameters
    ----------
    pdb_code : str
        4-letter valid PDB ID.
    chain_id : str
        Chain ID to analyze (e.g., 'A').
    ligand : str
        Optional: Name of ligand bound to PDB structure.
    max_retries : int
        Maximum number of retry attempts.
    initial_retry_delay : int
        Initial delay between retries in seconds.
    
    Returns
    -------
    str
        Job location URL.
    """
    for attempt in range(max_retries):
        try:
            r = requests.post(
                "https://proteins.plus/api/dogsite_rest",
                json={
                    "dogsite": {
                        "pdbCode": pdb_code,
                        "analysisDetail": "1",
                        "bindingSitePredictionGranularity": "1",
                        "ligand": ligand,
                        "chain": chain_id,
                    }
                },
                headers={"Content-type": "application/json", "Accept": "application/json"},
            )
            
            r.raise_for_status()
            return r.json()["location"]
        
        except HTTPError as e:
            if e.response.status_code == 429 and attempt < max_retries - 1:
                wait_time = initial_retry_delay * (2 ** attempt)  # Exponential backoff
                print(f"Rate limit exceeded. Waiting for {wait_time} seconds before retrying...")
                time.sleep(wait_time)
            else:
                raise

    raise Exception("Max retries exceeded")

def process_kinase_structures(kinase_name, max_structures=5):
    from opencadd.databases.klifs import setup_remote

    # Set up a KLIFS remote session
    klifs_session = setup_remote()

    try:
        structures = klifs_session.structures.by_kinase_name([kinase_name])
        print(f"Number of structures for {kinase_name}: {len(structures)}")
        print(structures.head())
        
        pdb_ids = structures["structure.pdb_id"].unique()[:max_structures]  # Limit to max_structures
        print(f"Processing {len(pdb_ids)} PDB IDs for {kinase_name}: {pdb_ids}")
        
        for pdb_id in pdb_ids:
            chain_id = structures[structures["structure.pdb_id"] == pdb_id]["structure.chain"].iloc[0]
            try:
                job_location = submit_dogsitescorer_job_with_pdbid(pdb_id, chain_id)
                print(f"DoGSiteScorer job for {kinase_name} (PDB {pdb_id}): {job_location}")
            except Exception as e:
                print(f"Failed to submit job for {kinase_name} (PDB {pdb_id}): {e}")
            
            # Add a longer delay between submissions to avoid rate limiting
            time.sleep(120)  # Wait for 2 minutes between submissions
        
    except ValueError as e:
        print(f"No structures found for {kinase_name}: {e}")

# List of kinases to investigate
kinase_names = ["CDK5", "DYRK1A"]

for kinase in kinase_names:
    process_kinase_structures(kinase)
    time.sleep(300)  # Wait for 5 minutes between kinases
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from opencadd.databases.klifs import setup_remote
import numpy as np
import pandas as pd

def get_klifs_pocket_residues(pdb_id):
    session = setup_remote()
    structures = session.structures.by_structure_pdb_id(pdb_id)
    structure_klifs_id = structures["structure.klifs_id"].iloc[0]
    klifs_pocket_df = session.pockets.by_structure_klifs_id(structure_klifs_id)
    
    # Clean the 'residue.id' column
    klifs_pocket_df['residue.id'] = klifs_pocket_df['residue.id'].apply(lambda x: ''.join(filter(str.isdigit, str(x))))
    klifs_pocket_df['residue.id'] = pd.to_numeric(klifs_pocket_df['residue.id'], errors='coerce')
    
    # Remove any rows with NaN values after conversion
    klifs_pocket_df = klifs_pocket_df.dropna(subset=['residue.id'])
    
    return klifs_pocket_df["residue.id"].astype(int).tolist()

def compare_pockets(dogsite_residues, klifs_residues, title):
    dogsite_set = set(dogsite_residues)
    klifs_set = set(klifs_residues)
    overlap = dogsite_set.intersection(klifs_set)
    
    print(f"{title}")
    print(f"DoGSiteScorer pocket with {len(dogsite_set)} residues detected.")
    print(f"KLIFS pocket with {len(klifs_set)} residues detected.")
    print(f"Overlap between the two residue sets: {len(overlap)}.")
    
    # Venn diagram visualization
    plt.figure(figsize=(10, 6))
    venn2([dogsite_set, klifs_set], set_labels=('DoGSiteScorer', 'KLIFS'))
    plt.title(title)
    plt.show()

# Perform comparison for DYRK1A structures
for pdb_id in dyrk1a_descriptors.keys():
    try:
        # Extract the siteAtms value and convert to a list of integers
        siteAtms = dyrk1a_descriptors[pdb_id][dyrk1a_descriptors[pdb_id]['name'] == 'P_0']['siteAtms'].iloc[0]
        if isinstance(siteAtms, np.int64):
            dyrk1a_dogsite_residues = [int(siteAtms)]
        else:
            dyrk1a_dogsite_residues = [int(x) for x in str(siteAtms).split(',') if x.strip() and x.isdigit()]
        
        dyrk1a_klifs_residues = get_klifs_pocket_residues(pdb_id)
        compare_pockets(dyrk1a_dogsite_residues, dyrk1a_klifs_residues, f"DYRK1A ({pdb_id}) Pocket Comparison")
    except Exception as e:
        print(f"Error processing {pdb_id}: {str(e)}")
```

# Kinase Binding Site Analysis: CDK5 and DYRK1A

## Introduction
This project aims to analyze and compare the binding sites of two important kinases involved in neurological diseases: Cyclin-Dependent Kinase 5 (CDK5) and Dual Specificity Tyrosine-Phosphorylation-Regulated Kinase 1A (DYRK1A). We employ computational methods to predict binding sites and compare them with expert-curated data to inform structure-based drug design efforts.

## Methodology

### 1. Data Collection
- Retrieve protein structures for CDK5 and DYRK1A from the Protein Data Bank (PDB).
- Access expert-curated binding site data from the KLIFS database.

### 2. Binding Site Prediction
- Use DoGSiteScorer to computationally predict binding sites.
- Extract binding site residues from KLIFS database.

### 3. Comparative Analysis
- Compare DoGSiteScorer predictions with KLIFS data for each structure.
- Visualize the overlap using Venn diagrams.

### 4. Cross-Kinase Comparison
- Compare binding site characteristics between CDK5 and DYRK1A.

## Scientific Rationale

### DoGSiteScorer
- Computational method for unbiased binding site prediction.
- Based on 3D protein structure analysis.

### KLIFS Database
- Expert-curated database specifically for kinases.
- Provides consistent binding site definitions across the kinome.

### Comparative Approach
- Allows validation of computational predictions against expert knowledge.
- Helps identify potential novel binding regions not captured in traditional analyses.

## Strengths and Weaknesses

### Strengths
1. Combines computational prediction with expert-curated data.
2. Provides a comprehensive view of kinase binding sites.
3. Allows for cross-kinase comparison.

### Weaknesses
1. Limited by the accuracy of computational prediction methods.
2. Reliance on static crystal structures may miss dynamic aspects of binding sites.
3. Discrepancy between DoGSiteScorer and KLIFS results requires careful interpretation.
