```python
# Cell 1: Imports and Data Loading
import pandas as pd
from tqdm.auto import tqdm
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

# Load the data from the previous step
egfr_data = pd.read_csv('../drug_design_2/ro5_properties_bace1_filtered.csv', index_col=0)
# Manually create RDKit molecules
egfr_data['ROMol'] = egfr_data['smiles'].apply(Chem.MolFromSmiles)

print(f"Initial number of compounds: {len(egfr_data)}")
```

```python
# Cell 2: Create Unwanted Substructures File
unwanted_substructures = [
    ("nitro", "[N+](=O)[O-]"),
    ("phosphate", "[P](=O)([OH])([OH])[OH]"),
    ("sulfate", "[S](=O)(=O)([OH])[OH]"),
    ("2-halopyridine", "c1([F,Cl,Br,I])ncccc1"),
    ("thiol", "[SH]"),
    ("alkyl_chain_7_or_more", "[CH2][CH2][CH2][CH2][CH2][CH2][CH2]"),
    ("aldehyde", "[CH]=O"),
    ("imine", "[CH]=[NH]"),
    ("quaternary_ammonium", "[N+]"),
    ("epoxide", "C1OC1"),
    ("sulfonyl_halide", "S(=O)(=O)[F,Cl,Br,I]"),
    ("beta_haloketone", "C(=O)C[F,Cl,Br,I]"),
    ("acyl_halide", "C(=O)[F,Cl,Br,I]"),
    ("michael_acceptor", "C=CC=O"),
]

unwanted_df = pd.DataFrame(unwanted_substructures, columns=['name', 'smarts'])
unwanted_df.to_csv('unwanted_substructures.csv', sep='\t', index=False)
print("Unwanted substructures file created with", len(unwanted_df), "entries")
```

```python
# Cell 3: Load and Process Unwanted Substructures
substructures = pd.read_csv("unwanted_substructures.csv", sep="\t")
substructures["rdkit_molecule"] = substructures.smarts.apply(Chem.MolFromSmarts)
print(f"Number of unwanted substructures: {len(substructures)}")
```

```python
# Cell 4: Visualize Some Unwanted Substructures (with error handling)
num_to_display = 3
valid_mols = []
valid_legends = []

for mol, legend in zip(substructures.rdkit_molecule.tolist()[:num_to_display], 
                       substructures.name.tolist()[:num_to_display]):
    try:
        img = Draw.MolToImage(mol)
        valid_mols.append(mol)
        valid_legends.append(legend)
    except Exception as e:
        print(f"Error visualizing {legend}: {str(e)}")

if valid_mols:
    img = Draw.MolsToGridImage(valid_mols, legends=valid_legends, subImgSize=(300, 300))
    display(img)
else:
    print("No valid molecules to display")
```

```python
# Cell 5: Filter for Unwanted Substructures
matches = []
clean = []

for index, row in tqdm(egfr_data.iterrows(), total=egfr_data.shape[0]):
    molecule = row.ROMol
    if molecule is None:
        continue
    match = False
    for _, substructure in substructures.iterrows():
        if molecule.HasSubstructMatch(substructure.rdkit_molecule):
            matches.append({
                "compound_id": index,  # Using index as identifier
                "rdkit_molecule": molecule,
                "substructure": substructure.rdkit_molecule,
                "substructure_name": substructure["name"],
            })
            match = True
            break
    if not match:
        clean.append(index)

matches = pd.DataFrame(matches)
egfr_data_filtered = egfr_data.loc[clean]

print(f"Number of compounds with unwanted substructures: {len(matches)}")
print(f"Number of compounds without unwanted substructures: {len(egfr_data_filtered)}")
```

```python
# Cell 6: Visualize Some Matches (with error handling)
def highlight_substructure(mol, sub):
    return mol.GetSubstructMatch(sub)

num_to_display = 9
valid_mols = []
valid_highlights = []
valid_legends = []

for _, row in matches.head(num_to_display).iterrows():
    try:
        highlight = highlight_substructure(row.rdkit_molecule, row.substructure)
        img = Draw.MolToImage(row.rdkit_molecule, highlightAtoms=highlight)
        valid_mols.append(row.rdkit_molecule)
        valid_highlights.append(highlight)
        valid_legends.append(f"{row.compound_id}: {row.substructure_name}")
    except Exception as e:
        print(f"Error visualizing match for {row.substructure_name}: {str(e)}")

if valid_mols:
    img = Draw.MolsToGridImage(valid_mols, highlightAtomLists=valid_highlights, 
                               legends=valid_legends, subImgSize=(300, 300))
    display(img)
else:
    print("No valid molecules to display")
```

```python
# Cell 7: Substructure Statistics
group_frequencies = matches.groupby("substructure_name").size().sort_values(ascending=False)
print("Top 10 most frequent unwanted substructures:")
print(group_frequencies.head(10))
```

```python
# Cell 8: Save Filtered Dataset
egfr_data_filtered.to_csv('bace1_filtered_final.csv')
print(f"Filtered dataset saved with {len(egfr_data_filtered)} compounds.")
```

```python
# Cell 9: Calculate Removal Percentage
initial_count = len(egfr_data)
final_count = len(egfr_data_filtered)
removal_percentage = (initial_count - final_count) / initial_count * 100
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem

# Define the substructures and their SMARTS patterns
substructures = [
    ("2-halopyridine", "c1([F,Cl,Br,I])ncccc1", "Highly reactive, can lead to covalent binding with proteins"),
    ("michael_acceptor", "C=CC=O", "Can form covalent bonds with proteins, potentially leading to toxicity"),
    ("nitro", "[N+](=O)[O-]", "Can be reduced to reactive intermediates, potentially causing DNA damage"),
    ("quaternary_ammonium", "[N+]", "Can interfere with ion channels and neurotransmitter receptors"),
    ("beta_haloketone", "C(=O)C[F,Cl,Br,I]", "Highly reactive, can form covalent bonds with proteins"),
    ("alkyl_chain_7_or_more", "[CH2][CH2][CH2][CH2][CH2][CH2][CH2]", "Can lead to poor solubility and increased lipophilicity"),
    ("aldehyde", "[CH]=O", "Highly reactive, can form Schiff bases with amines in proteins"),
    ("epoxide", "C1OC1", "Can alkylate DNA and proteins, potentially leading to mutagenicity"),
    ("thiol", "[SH]", "Can form disulfide bonds and interfere with protein structure")
]

# Create molecules from SMARTS
mols = []
legends = []
explanations = []

for name, smarts, explanation in substructures:
    mol = Chem.MolFromSmarts(smarts)
    if mol is not None:
        Chem.AllChem.Compute2DCoords(mol)
        mols.append(mol)
        legends.append(name)
        explanations.append(f"{name}: {explanation}")

# Generate the image
img = Draw.MolsToGridImage(mols, molsPerRow=3, subImgSize=(300,300), legends=legends)

# Display the image and explanations
display(img)
print("\nExplanations:")
for explanation in explanations:
    print(f"- {explanation}")

print("\nNote: These substructures are generally considered problematic, but their presence doesn't always guarantee toxicity. Context and specific molecular environment are important factors.")
```

# Drug Discovery Process: Filtering Unwanted Substructures

## Overview
In this notebook, we've implemented a crucial step in the drug discovery pipeline: filtering out compounds with potentially problematic substructures. This process is essential for improving the quality of our compound library and increasing the likelihood of identifying safe and effective drug candidates.

## Steps and Scientific Background

### 1. Data Loading and Preparation
We started by loading our BACE1 (Beta-secretase 1) inhibitor dataset, which had already been filtered for Lipinski's Rule of 5. This initial dataset represents potential drug candidates for Alzheimer's disease treatment.

BACE1 is an enzyme involved in the production of amyloid-beta peptides, which are associated with Alzheimer's disease. Inhibiting BACE1 is a potential therapeutic strategy.

### 2. Defining Unwanted Substructures
We created a list of unwanted substructures based on known problematic chemical groups in drug discovery.

These substructures are often associated with toxicity, reactivity, or poor pharmacokinetic properties:
- 2-halopyridines: Highly reactive, can lead to covalent binding with proteins
- Michael acceptors: Can form covalent bonds with proteins
- Nitro groups: Can be reduced to reactive intermediates
- Quaternary ammonium compounds: Can interfere with ion channels
- Beta-haloketones: Highly reactive, can form covalent bonds with proteins
- Long alkyl chains: Can lead to poor solubility and increased lipophilicity
- Aldehydes: Can form Schiff bases with protein amines
- Epoxides: Can alkylate DNA and proteins
- Thiols: Can form disulfide bonds and interfere with protein structure

### 3. Substructure Filtering
We used RDKit to identify and filter out compounds containing these unwanted substructures.

This step helps to reduce the risk of pursuing compounds that may later fail due to toxicity or other issues. It's a form of negative design in drug discovery.

### 4. Visualization and Analysis
We visualized some of the filtered compounds and their problematic substructures, and analyzed the frequency of each unwanted substructure in our dataset.

This helps us understand the composition of our compound library and identify which problematic features are most common.

## Results
We successfully filtered our compound library, removing molecules with potentially problematic substructures. This resulted in a smaller, but potentially higher-quality set of compounds for further investigation.
