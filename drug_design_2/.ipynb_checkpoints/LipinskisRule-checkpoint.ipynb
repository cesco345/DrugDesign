```python
# Cell 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors, PandasTools
import math
```

```python
# Cell 2: Load the BACE1 dataset (Updated)
molecules = pd.read_csv("../drug_design_1/BACE1_compounds.csv")
print(molecules.shape)
molecules.head()

```

```python
# Cell 3: Define function to calculate Ro5 properties
def calculate_ro5_properties(smiles):
    molecule = Chem.MolFromSmiles(smiles)
    molecular_weight = Descriptors.ExactMolWt(molecule)
    n_hba = Descriptors.NumHAcceptors(molecule)
    n_hbd = Descriptors.NumHDonors(molecule)
    logp = Descriptors.MolLogP(molecule)
    conditions = [molecular_weight <= 500, n_hba <= 10, n_hbd <= 5, logp <= 5]
    ro5_fulfilled = sum(conditions) >= 3
    return pd.Series(
        [molecular_weight, n_hba, n_hbd, logp, ro5_fulfilled],
        index=["molecular_weight", "n_hba", "n_hbd", "logp", "ro5_fulfilled"],
    )
```

```python
# Cell 4: Perform Ro5 calculations on all compounds
ro5_properties = molecules["smiles"].apply(calculate_ro5_properties)
```

```python
# Cell 5: Concatenate molecules with Ro5 data calculations
molecules = pd.concat([molecules, ro5_properties], axis=1)
```

```python
# Cell 6: Filter compounds based on Ro5 
molecules_ro5_fulfilled = molecules[molecules["ro5_fulfilled"] == True].copy()
molecules_ro5_violated = molecules[molecules["ro5_fulfilled"] == False].copy()

print(f"# compounds in unfiltered data set: {molecules.shape[0]}")
print(f"# compounds in filtered data set: {molecules_ro5_fulfilled.shape[0]}")
print(f"# compounds not compliant with the Ro5: {molecules_ro5_violated.shape[0]}")
```

```python
# Cell 7: Save the datasets 
molecules.to_csv('ro5_properties_bace1.csv', index=False)
molecules_ro5_fulfilled.to_csv('ro5_properties_bace1_filtered.csv', index=False)
```

```python
# Cell 8: Helper functions for radar plot
def calculate_mean_std(dataframe):
    stats = dataframe.describe().T[["mean", "std"]]
    return stats
def _scale_by_thresholds(stats, thresholds, scaled_threshold):
    for property_name in stats.index:
        if property_name not in thresholds.keys():
            raise KeyError(f"Add property '{property_name}' to scaling variable.")
    stats_scaled = stats.apply(lambda x: x / thresholds[x.name] * scaled_threshold, axis=1)
    return stats_scaled

def _define_radial_axes_angles(n_axes):
    x_angles = [i / float(n_axes) * 2 * math.pi for i in range(n_axes)]
    x_angles += x_angles[:1]
    return x_angles

def plot_radar(y, thresholds, scaled_threshold, properties_labels, y_max=None, output_path=None):
    x = _define_radial_axes_angles(len(y))
    y = _scale_by_thresholds(y, thresholds, scaled_threshold)
    y = pd.concat([y, y.head(1)])

    plt.figure(figsize=(6, 6))
    ax = plt.subplot(111, polar=True)

    ax.fill(x, [scaled_threshold] * len(x), "cornflowerblue", alpha=0.2)
    ax.plot(x, y["mean"], "b", lw=3, ls="-")
    ax.plot(x, y["mean"] + y["std"], "orange", lw=2, ls="--")
    ax.plot(x, y["mean"] - y["std"], "orange", lw=2, ls="-.")

    ax.set_theta_offset(math.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_rlabel_position(180)
    plt.xticks(x, [])
    if not y_max:
        y_max = int(ax.get_yticks()[-1])
    plt.ylim(0, y_max)
    plt.yticks(
        range(1, y_max),
        ["5" if i == scaled_threshold else "" for i in range(1, y_max)],
        fontsize=16,
    )

    for i, (angle, label) in enumerate(zip(x[:-1], properties_labels)):
        if angle == 0:
            ha = "center"
        elif 0 < angle < math.pi:
            ha = "left"
        elif angle == math.pi:
            ha = "center"
        else:
            ha = "right"
        ax.text(
            x=angle,
            y=y_max + 1,
            s=label,
            size=16,
            horizontalalignment=ha,
            verticalalignment="center",
        )

    labels = ("mean", "mean + std", "mean - std", "rule of five area")
    ax.legend(labels, loc=(1.1, 0.7), labelspacing=0.3, fontsize=16)

    if output_path:
        plt.savefig(output_path, dpi=300, bbox_inches="tight", transparent=True)

    plt.show()
```

```python
# Cell 9: Calculate statistics for filtered and violated datasets
molecules_ro5_fulfilled_stats = calculate_mean_std(
    molecules_ro5_fulfilled[["molecular_weight", "n_hba", "n_hbd", "logp"]]
)
molecules_ro5_violated_stats = calculate_mean_std(
    molecules_ro5_violated[["molecular_weight", "n_hba", "n_hbd", "logp"]]
)
```

```python
# Cell 10: Define parameters for radar plots
thresholds = {"molecular_weight": 500, "n_hba": 10, "n_hbd": 5, "logp": 5}
scaled_threshold = 5
properties_labels = [
    "Molecular weight (Da) / 100",
    "# HBA / 2",
    "# HBD",
    "LogP",
]
y_max = 8
```

```python
# Cell 11: Generate radar plot for Ro5-compliant compounds
plot_radar(
    molecules_ro5_fulfilled_stats,
    thresholds,
    scaled_threshold,
    properties_labels,
    y_max,
    output_path="bace1_ro5_fulfilled_radar.png"
)
```

```python
# Cell 12: Generate radar plot for Ro5-violated compounds
plot_radar(
    molecules_ro5_violated_stats,
    thresholds,
    scaled_threshold,
    properties_labels,
    y_max,
    output_path="bace1_ro5_violated_radar.png"
)
```

```python
# Cell 13: Inspect DataFrame columns
print(molecules.columns)
```

```python
# Function to calculate drug-likeness score
def calculate_druglikeness_score(row):
    score = 0
    if row['molecular_weight'] <= 500:
        score += 1
    if row['n_hba'] <= 10:
        score += 1
    if row['n_hbd'] <= 5:
        score += 1
    if row['logp'] <= 5:
        score += 1
    return score

# Calculate drug-likeness score for all molecules
molecules['druglikeness_score'] = molecules.apply(calculate_druglikeness_score, axis=1)

# Refilter the molecules with the new score
molecules_ro5_fulfilled = molecules[molecules["ro5_fulfilled"] == True].copy()
molecules_ro5_violated = molecules[molecules["ro5_fulfilled"] == False].copy()

# Sort compliant drugs by druglikeness score (higher is better)
top_10_compliant = molecules_ro5_fulfilled.sort_values('druglikeness_score', ascending=False).head(10)

# Sort non-compliant drugs by druglikeness score (higher is better)
top_10_violated = molecules_ro5_violated.sort_values('druglikeness_score', ascending=False).head(10)

# Use 'molecule_chembl_id' as the identifier column
identifier_column = 'molecule_chembl_id'

print("Top 10 Ro5 Compliant Drugs:")
for i, (index, row) in enumerate(top_10_compliant.iterrows(), 1):
    print(f"{i}. {row[identifier_column]} (Score: {row['druglikeness_score']}, pIC50: {row['pIC50']:.2f})")

print("\nTop 10 Ro5 Non-Compliant Drugs:")
for i, (index, row) in enumerate(top_10_violated.iterrows(), 1):
    print(f"{i}. {row[identifier_column]} (Score: {row['druglikeness_score']}, pIC50: {row['pIC50']:.2f})")
```

```python
# Cell 15: Visualize top 10 compliant and non-compliant compounds

from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
import matplotlib.pyplot as plt

def draw_molecules(mol_list, titles, n_cols=5, mol_size=(300, 300)):
    n_rows = (len(mol_list) - 1) // n_cols + 1
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(4*n_cols, 4*n_rows))
    axes = axes.flatten()
    
    for i, (mol, title) in enumerate(zip(mol_list, titles)):
        img = Draw.MolToImage(mol, size=mol_size)
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(title)
    
    for j in range(i+1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

# Function to get molecules and titles
def get_mols_and_titles(df):
    mols = []
    titles = []
    for _, row in df.iterrows():
        mol = Chem.MolFromSmiles(row['smiles'])
        if mol is not None:
            AllChem.Compute2DCoords(mol)
            mols.append(mol)
            titles.append(f"{row['molecule_chembl_id']}\npIC50: {row['pIC50']:.2f}")
    return mols, titles

# Draw compliant compounds
compliant_mols, compliant_titles = get_mols_and_titles(top_10_compliant)
print("Top 10 Ro5 Compliant Compounds:")
draw_molecules(compliant_mols, compliant_titles)

# Draw non-compliant compounds
non_compliant_mols, non_compliant_titles = get_mols_and_titles(top_10_violated)
print("Top 10 Ro5 Non-Compliant Compounds:")
draw_molecules(non_compliant_mols, non_compliant_titles)
```

```python
# Cell 16: Visualization for CHEMBL3897461

from rdkit import Chem
from rdkit.Chem import AllChem, Draw
from rdkit.Chem import rdDepictor
import py3Dmol

# Get the SMILES for CHEMBL3897461
chembl3897461_data = molecules[molecules['molecule_chembl_id'] == 'CHEMBL3897461'].iloc[0]
smiles = chembl3897461_data['smiles']

# Create RDKit molecule
mol = Chem.MolFromSmiles(smiles)

# Generate 2D coordinates
rdDepictor.Compute2DCoords(mol)

# Draw the molecule
img = Draw.MolToImage(mol, size=(400, 400))
display(img)

# Print molecule details
print(f"CHEMBL ID: {chembl3897461_data['molecule_chembl_id']}")
print(f"SMILES: {smiles}")
print(f"Molecular Weight: {chembl3897461_data['molecular_weight']:.2f}")
print(f"LogP: {chembl3897461_data['logp']:.2f}")
print(f"H-Bond Donors: {chembl3897461_data['n_hbd']}")
print(f"H-Bond Acceptors: {chembl3897461_data['n_hba']}")
print(f"pIC50: {chembl3897461_data['pIC50']:.2f}")

# Generate 3D coordinates for a more detailed view
mol_3d = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol_3d, randomSeed=42)
AllChem.MMFFOptimizeMolecule(mol_3d)

# Create a 3D viewer
viewer = py3Dmol.view(width=400, height=400)
viewer.addModel(Chem.MolToMolBlock(mol_3d), "mol")
viewer.setStyle({'stick': {}})
viewer.zoomTo()
viewer.render()
display(viewer)
```

# BACE1 Inhibitor Analysis: Summary, Scientific Background, and Next Steps

## Overview
In this notebook, we analyzed a dataset of BACE1 (Beta-Secretase 1) inhibitors, focusing on their drug-likeness properties and potential as candidates for Alzheimer's disease treatment. We applied Lipinski's Rule of Five (Ro5) to assess the compounds' likelihood of being orally active drugs in humans.

## Steps Taken and Scientific Rationale

1. **Data Loading and Preparation**
   - We loaded the BACE1 inhibitor dataset, which contained molecular structures (as SMILES) and activity data (pIC50).
   - BACE1 is a key enzyme in the formation of amyloid-β plaques, a hallmark of Alzheimer's disease. Inhibiting BACE1 is a promising therapeutic strategy.

2. **Calculation of Molecular Properties**
   - We used RDKit to calculate key molecular properties: molecular weight, number of hydrogen bond donors/acceptors, and LogP.
   - These properties are crucial in determining a compound's drug-likeness and potential for oral bioavailability.

3. **Application of Lipinski's Rule of Five**
   - We assessed each compound against Ro5 criteria: MW ≤ 500, HBD ≤ 5, HBA ≤ 10, LogP ≤ 5.
   - Ro5 is a widely used heuristic to predict the likelihood of a compound being orally active in humans. Compounds that follow these rules are more likely to have good absorption and permeation.

4. **Druglikeness Scoring**
   - We created a simple scoring system based on Ro5 compliance.
   - This allows for a nuanced view of drug-likeness, recognizing that some successful drugs violate one or more of Lipinski's rules.

5. **Analysis of Compliant vs. Non-Compliant Compounds**
   - We separated and analyzed Ro5-compliant and non-compliant compounds.
   - This comparison helps identify potential lead compounds and understand the trade-offs between optimal BACE1 inhibition and drug-like properties.

6. **Visualization of Top Compounds**
   - We generated 2D and 3D visualizations of top-scoring compounds.
   - Visual inspection of molecular structures can reveal important structural features that contribute to both activity and drug-likeness.

## Key Findings
- We identified top compounds that balance good BACE1 inhibition (high pIC50) with favorable drug-like properties.
- Some highly potent BACE1 inhibitors violate Ro5, suggesting a potential trade-off between activity and oral bioavailability.
- Visual analysis revealed structural motifs that might be important for BACE1 inhibition or contribute to Ro5 violations.
